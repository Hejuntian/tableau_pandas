```python
# 把df 的栏位用列表推导的方式清理掉'年'这个字符後转换成整数
df.columns = [ int(x.replace('年','')) if x!='地区' else x for x in df.columns]
df.columns
```

```python
# 求最新的一年
years = [ x for x in df.columns if x!='地区'] 
max(years)
```

```python
# pd.melt() 相当於Tableau 的数据透视
df_melted = pd.melt(df, id_vars=['地区'], value_vars=years)
```

```python
df_Local_done = df_melted
```

```python
df = df_Tour
df.columns = [ int(x.replace('年','')) if x!='地区' else x for x in df.columns]
years = [ x for x in df.columns if x!='地区'] 
df_melted = pd.melt(df, id_vars=['地区'], value_vars=years)
df_melted.columns = ['地区', '年', '国际旅游外汇收入（百万美元）']
df_Tour_done = df_melted
df_Tour_done
```

```python
df_p = df_Local_done.join(df_Tour_done,lsuffix='', rsuffix='_Tour')
df_p
```

```python
df_pp = df_p.drop(["地区_Tour","年_Tour"], axis=1).set_index(["地区","年"])
```

```python
# 解决中文乱码问题
import matplotlib as mpl
import matplotlib.pyplot as plt
import pylab
mpl.rcParams['font.sans-serif']=['SimHei']
mpl.rcParams['axes.unicode_minus']=False

df_plot = df_pp.query("年==2015")
df_plot.plot(kind='scatter',x='国际旅游外汇收入（百万美元）', y='财政收入（亿元）', 
          title = '国际旅游外汇收入（百万美元）与财政收入（亿元）的关系：分省数据')
```

```python
# Regression Using Pandas and Statsmodels
import numpy as np
fit = np.polyfit(x=df_pp.query("年==2015")['国际旅游外汇收入（百万美元）'], y=df_pp.query("年==2015")['财政收入（亿元）'], deg=1)
```

```python
x=df_plot['国际旅游外汇收入（百万美元）']
y=df_plot['财政收入（亿元）']
ax = df_plot.plot(kind='scatter', x='国际旅游外汇收入（百万美元）', y='财政收入（亿元）', 
          title = '国际旅游外汇收入（百万美元）与财政收入（亿元）的关系：分省数据')
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y)
```

```python
# 制图
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.models import HoverTool, BoxSelectTool

title = '分省数据' 
xlabel = '国际旅游外汇收入（百万美元）' 
ylabel = '财政收入（亿元）'

hover = HoverTool(
    tooltips=[
        ( '国际旅游外汇收入（百万美元）',  '$x{0,F}'), 
        ( '财政收入（亿元）',  '$y{0,F}'),
        ( '地区',  '@desc'  ),
    ]
)

TOOLS = [hover, "pan, wheel_zoom, box_zoom, reset, resize"]
fig = figure(plot_width=800, plot_height=600,
           x_axis_label = xlabel, y_axis_label = ylabel, 
           title=title, tools=TOOLS)

df_plot = df_pp.query("年==2015")

fig.circle("国际旅游外汇收入（百万美元）", "财政收入（亿元）",
         source = df_plot,
         fill_alpha=0.6, line_color=None, size= 30)

#regression line
xs = [min(df_plot["国际旅游外汇收入（百万美元）"]), max(df_plot["国际旅游外汇收入（百万美元）"])]
fig.line(x=xs, y=[ fit[0] * x + fit[1] for x in xs] )

output_notebook() # 输出到IPython/Jupyter Notebook，若要输出到他处需调整
#output_file('by_provinces.htm') # 输出到 htm
show(fig)
```

```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, push_notebook, show
from bokeh.models import CustomJS, Slider,HoverTool, BoxSelectTool
from bokeh.layouts import column

title = '分省数据' 
xlabel = '国际旅游外汇收入（百万美元）' 
ylabel = '财政收入（亿元）'
hover = HoverTool(
    tooltips=[
        ( '国际旅游外汇收入（百万美元）',  '$x{0,F}'), 
        ( '财政收入（亿元）',  '$y{0,F}'),
        ( '地区',  '@desc'  ),
    ]
)

TOOLS = [hover, "pan, wheel_zoom, box_zoom, reset, resize"]

fig = figure(plot_width=800, plot_height=600,
           x_axis_label = xlabel, y_axis_label = ylabel, 
           title=title, tools=TOOLS)

yr_min = df_pp.index.get_level_values(1).min()
df_plot = df_pp.query("年=={yr}".format(yr=yr_min))
plt = fig.circle("国际旅游外汇收入（百万美元）", "财政收入（亿元）",
         source = df_plot,
         fill_alpha=0.6, line_color=None, size= 15)


def update_plot(year):
    #plt.data_source.data = df_pp.query("年=={yr}".format(yr=year))
    plt.data_source.data['国际旅游外汇收入（百万美元）'] = list(df_pp.query("年=={yr}".format(yr=year)).loc[:,['国际旅游外汇收入（百万美元）']].iloc[:,0].values)
    plt.data_source.data['财政收入（亿元）'] = list(df_pp.query("年=={yr}".format(yr=year)).loc[:,['财政收入（亿元）']].iloc[:,0].values)
    plt.data_source.data['地区_年'] = [(x[0],int(year)) for x in plt.data_source.data['地区_年']]

    push_notebook(handle=bokeh_handle)  
    print (year)
    ##### new notebook cell #####

callback = CustomJS(code="""
if (IPython.notebook.kernel !== undefined) {
    var kernel = IPython.notebook.kernel;
    cmd = "update_plot(" + cb_obj.value + ")";
    kernel.execute(cmd, {}, {});
}
""")

slider = Slider(start=df_pp.index.get_level_values(1).min(), 
                end=df_pp.index.get_level_values(1).max(),
                value=1,
                step=1,
                title="年",
                callback=callback)
output_notebook()

bokeh_handle = show(column(slider, fig), notebook_handle=True)
```

```python
yr = 2015
df_plot = df_pp.query("年=={yr}".format(yr=yr_min))
# Regression Using Pandas and Statsmodels
fit = np.polyfit(x=df_plot['国际旅游外汇收入（百万美元）'], 
                 y=df_plot['财政收入（亿元）'], deg=1)
```

```python
df_plot['expected'] =  [ fit[0] * x + fit[1] for x in df_plot['国际旅游外汇收入（百万美元）']] 
df_plot['difference'] =  df_plot['财政收入（亿元）'] - df_plot['expected']
df_plot['difference_ratio'] = ( df_plot['财政收入（亿元）'] - df_plot['expected'] ) /df_plot['国际旅游外汇收入（百万美元）'] 
```

```python
df_plot.sort_values(by='difference_ratio')
```
